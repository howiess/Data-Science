# Commonly Used Supervised Learning Algorithms

### Here are 10 most commonly used supervised machine learning algorithms in the industry 


## 1. Linear Regression
- **Usage**: Predicting continuous numerical outcomes.
- **Applications**: House price prediction, stock price forecasting, risk management.
- **Additional Point**: Understand assumptions like linearity, independence, homoscedasticity, and normality of errors. Recognize how to diagnose and handle multicollinearity.

- **Key Interview Prep**: Explain the interpretation of model coefficients and odds ratios. Understand the confusion matrix, precision, recall, F1 score, and ROC curves.

## 3. Decision Tree
- **Usage**: Classification and regression.
- **Applications**: Customer segmentation, credit scoring, medical diagnosis.
- **Additional Point**: Understand overfitting in decision trees and how to mitigate it using techniques like pruning, setting a maximum depth, or minimum samples per leaf.

## 4. Random Forest
- **Usage**: Ensemble method for classification and regression.

- **Additional Point**: Understand feature importance and how Random Forest can be used for feature selection. Learn about out-of-bag (OOB) error estimation.
- **Key Interview Prep**: Discuss the trade-offs between model interpretability and performance. Explain how ensemble methods reduce variance and improve robustness.

## 5. Support Vector Machines (SVM)
- **Usage**: Classification and regression.
- **Applications**: Image classification, text categorization, handwriting recognition.
- **Additional Point**: Learn about different kernel functions (linear, polynomial, RBF) and their applications. Understand the impact of the choice of kernel and hyperparameters on the model performance.
- **Key Interview Prep**: Be able to explain margin maximization and the concept of support vectors. Understand the trade-off parameter (C) and how it affects the decision boundary.

## 6. K-Nearest Neighbors (KNN)

- **Applications**: Recommender systems, anomaly detection, image recognition.
- **Additional Point**: Understand the computational complexity of KNN and techniques to make it more efficient, such as KD-trees or Ball-trees.
- **Key Interview Prep**: Discuss the impact of the choice of distance metric and the number of neighbors (k) on model performance. Understand scenarios where KNN is suitable despite its simplicity.

## 7. Naive Bayes
- **Usage**: Probabilistic classification.
- **Applications**: Text classification, spam filtering, sentiment analysis.
- **Additional Point**: Learn about different types of Naive Bayes classifiers (Gaussian, Multinomial, Bernoulli) and their applications based on data distribution.
- **Key Interview Prep**: Explain how Naive Bayes handles missing data and categorical variables. Understand how it can be surprisingly effective despite its simplistic assumptions.

## 8. Gradient Boosting Machines (GBM)
- **Usage**: Ensemble method for classification and regression.
- **Applications**: Predictive maintenance, marketing response prediction, sales forecasting.
- **Additional Point**: Understand the concept of boosting and how it differs from bagging. Learn about the learning rate, n_estimators, and how these parameters affect model performance.

## 9.XGBoost
- **Usage**: Optimized gradient boosting.
- **Applications**: Kaggle competitions, real-time prediction systems, personalized recommendations.
- **Additional Point**: Understand the optimization techniques XGBoost uses, such as tree pruning and handling of missing values. Learn about advanced regularization techniques like L1 and L2.
- **Key Interview Prep**: Discuss the benefits of XGBoost over traditional gradient boosting. Explain hyperparameter tuning and cross-validation strategies for model improvement.

## 10. AdaBoost
- **Usage**: Ensemble method for boosting weak classifiers.
- **Applications**: Face detection, text classification, fraud detection.
- **Additional Point**: Understand how AdaBoost combines multiple weak classifiers to create a strong classifier. Learn about the impact of misclassified instances on the model.
- **Key Interview Prep**: Be prepared to explain the boosting process, the role of weights in AdaBoost, and how the algorithm adapts to difficult instances.

## 1. Linear Regression

<h2>Step 1: Understanding Linear Regression</h2>
<p>Linear Regression is a method to model the relationship between a dependent variable <em>y</em> and one or more independent variables <em>X</em>. The relationship is modeled as:</p>

$$y = β_{0} + β_{1}X_{1} + β_{2}X_{2}+ ... + β_{n}X_{n} + \sigma$$

<p>Where:</p>
<ul>
<li><em>y</em> is the dependent variable.</li>
<li><em>β<sub>0</sub></em> is the y-intercept.</li>
<li><em>β<sub>1</sub>, β<sub>2</sub>, ⋯, β<sub>n</sub></em> are the coefficients.</li>
<li><em>X<sub>1</sub>, X<sub>2</sub>, ⋯, X<sub>n</sub></em> are the independent variables.</li>
<li><em>ε</em> is the error term.</li>
</ul>

<h2>Step 2: Assessing Model Assumptions</h2>
<p>Before implementing the algorithm, ensure the following assumptions hold:</p>

<h3>Linearity</h3>
<p>The relationship between the independent variables and the dependent variable is linear.</p>
<p><strong>Test</strong>: Scatter plots of each independent variable against the dependent variable.</p>

<h3>Independence</h3>
<p>The observations are independent of each other.</p>
<p><strong>Test</strong>: Durbin-Watson test to check for autocorrelation in residuals.</p>

<h3>Homoscedasticity</h3>
<p>The residuals (errors) have constant variance.</p>
<p><strong>Test</strong>: Plot residuals vs. fitted values. Look for patterns indicating non-constant variance.</p>

<h3>Normality</h3>
<p>The residuals of the model are normally distributed.</p>
<p><strong>Test</strong>: Q-Q plot of residuals against a normal distribution.</p>

<h3>No Multicollinearity</h3>
<p>The independent variables are not too highly correlated.</p>
<p><strong>Test</strong>: Calculate Variance Inflation Factor (VIF) for each independent variable.</p>

<h2>Step 3: Data Preparation</h2>
<ol>
<li><strong>Collect Data</strong>: Gather the dataset you will be working with.</li>
<li><strong>Clean Data</strong>: Handle missing values, remove duplicates, and perform any necessary preprocessing.</li>
<li><strong>Split Data</strong>: Divide the data into training and testing sets.</li>
</ol>

<h2>Step 4: Mathematical Concepts</h2>
<p><strong>Ordinary Least Squares (OLS)</strong>: The method to estimate the coefficients <em>β</em> in Linear Regression. The goal is to minimize the sum of the squared residuals:</p>

$$RSS=\sum{(y_{i}-\hat{y}_{i})^{2}}$$

<p>Where <em>ŷ<sub>i</sub></em> is the predicted value.</p>

<p>The formula to calculate the coefficients using OLS is:</p>

$$\beta=(X^{T}X)^{-1}X^{T}y$$

<p>Where:</p>
<ul>
<li><em>X</em> is the matrix of input features.</li>
<li><em>y</em> is the vector of output values.</li>
<li><em>β</em> is the vector of coefficients.</li>
</ul>
<p>It is derived by:<p>

$$RSS = (y-\hat{y})^T(y-\hat{y}) \text{\ \ \ in matrix form}$$
$$= (y-X\beta)^T(y-X\beta)$$
$$= y^Ty-y^TX\beta-(X\beta)^Ty+(X\beta)^T(X\beta)$$ 
$$= y^Ty-y^TX\beta-\beta^TX^Ty+\beta^TX^TX\beta \text{\ \ \ <= $y^TX\beta$ and $\beta^TX^Ty$ are scalars and are equal to each other}$$
$$= y^Ty-2y^TX\beta+\beta^TX^TX\beta$$
<h2>Step 5: Implementation from Scratch</h2>

<h3>Add Intercept Term</h3>
<p>Add a column of ones to the input features to account for the intercept term.</p>

<p><strong>X<sub>b</sub> = [1 X<sub>11</sub> X<sub>12</sub> ⋯ X<sub>1n</sub>;<br>1 X<sub>21</sub> X<sub>22</sub> ⋯ X<sub>2n</sub>;<br>⋮ ⋮ ⋮ ⋱ ⋮;<br>1 X<sub>m1</sub> X<sub>m2</sub> ⋯ X<sub>mn</sub>]</strong></p>

<h3>Calculate Coefficients</h3>
<p>Use the OLS formula to calculate the coefficients.</p>

<p><strong>β = (X<sub>b</sub><sup>T</sup> X<sub>b</sub>)<sup>-1</sup> X<sub>b</sub><sup>T</sup> y</strong></p>

<h3>Make Predictions</h3>
<p>Use the calculated coefficients to make predictions.</p>

<p><strong>ŷ = X<sub>b</sub> β</strong></p>

<h3>Evaluate the Model</h3>
<p>To evaluate the model, calculate metrics such as Mean Squared Error (MSE):</p>

$$MSE=\frac{1}{n}\sum{(y_{i}-\hat{y}_{i})^{2}}$$

<h2>Step-by-Step Summary</h2>
<ol>
<li><strong>Understand Linear Regression</strong>: Recognize the relationship and the formula.</li>
<li><strong>Assess Model Assumptions</strong>: Ensure the assumptions of linearity, independence, homoscedasticity, normality, and no multicollinearity.</li>
<ul>
<li><strong>Linearity Test</strong>: Use scatter plots.</li>
<li><strong>Independence Test</strong>: Durbin-Watson test.</li>
<li><strong>Homoscedasticity Test</strong>: Plot residuals vs. fitted values.</li>
<li><strong>Normality Test</strong>: Q-Q plot.</li>
<li><strong>Multicollinearity Test</strong>: Calculate VIF.</li>
</ul>
<li><strong>Prepare Data</strong>: Clean and split your data.</li>
<li><strong>Mathematical Concept - OLS</strong>: Understand and apply the OLS formula to estimate coefficients.</li>
<li><strong>Implement Algorithm</strong>:
<ul>
<li>Add intercept term.</li>
<li>Calculate coefficients using the OLS formula.</li>
<li>Make predictions.</li>
</ul>
</li>

</ol>


## 2. Logistic Regression


<h2>Overview</h2>
<p>Logistic Regression is a statistical method used for binary classification problems, where the outcome can take one of two possible values. It models the probability that a given input belongs to a particular class.</p>

<h2>Mathematical Concepts</h2>

<h3>1. Logistic Function (Sigmoid Function)</h3>
<p>The logistic function maps any real-valued number into the range [0, 1].</p>
<p><strong>Formula:</strong></p>

$$\sigma(z)=\frac{1}{1+e^{-z}}$$

<p>Where z is the linear combination of input features.</p>

<h3>2. Model Equation</h3>
<p>The linear combination of inputs is passed through the logistic function.</p>
<p><strong>Formula:</strong></p>

$$\hat{p}=\sigma(\beta_{0}+\beta_{1}X_{1}+\beta_{2}X_{2}+...+\beta_{n}X_{n})$$

<h3>3. Log-Odds and Odds</h3>
<p><strong>Odds:</strong> The ratio of the probability of the event occurring to the probability of it not occurring.</p>

$$Odds = \frac{p}{1-p}$$

<p><strong>Log-Odds (Logit):</strong> The natural logarithm of the odds.</p>

$$Odds = log\frac{p}{1-p}$$

<p>In logistic regression:</p>

$$Logit(p) = \beta_{0}+\beta_{1}X_{1}+\beta_{2}X_{2}+...+\beta_{n}X_{n}$$

<h3>4. Cost Function</h3>
<p>Logistic regression uses the log loss (binary cross-entropy) as the cost function.</p>
<p><strong>Formula:</strong></p>

$$J(\beta)=-\frac{1}{m}\sum^{m}_{i=1}[y_{i}log(\hat{p}_{i})+(1-y_{i})log(1-\hat{p}_{i})]$$

<h3>5. Gradient Descent</h3>
<p>To minimize the cost function, gradient descent is used.</p>
<p><strong>Gradient of the Cost Function:</strong></p>

$$\frac{dJ(\beta)}{d\beta_{j}}=\frac{1}{m}\sum^{m}_{i=1}(\hat{p}_{i}-y_{i})X_{ij}$$

<p><strong>Update Rule:</strong></p>

$$\beta_{j} := \beta_{j}-\alpha\frac{dJ(\beta)}{d\beta_{j}}$$

<p>Where &alpha; is the learning rate.</p>

<h2>Assumptions of Logistic Regression</h2>
<ul>
<li><strong>Linearity in Logit:</strong> The logit (log-odds) of the outcome is a linear combination of the independent variables.</li>
<li><strong>Independent Observations:</strong> Observations are independent of each other.</li>
<li><strong>No Multicollinearity:</strong> Independent variables are not too highly correlated with each other.</li>
<li><strong>Large Sample Size:</strong> Logistic regression requires a large sample size to provide accurate estimates.</li>
</ul>

<h2>Things to Be Mindful Of</h2>
<ul>
<li><strong>Outliers:</strong> Logistic regression can be sensitive to outliers.</li>
<li><strong>Imbalanced Data:</strong> If the classes are imbalanced, the model may perform poorly on the minority class.</li>
<li><strong>Feature Scaling:</strong> Features should be scaled for better convergence of gradient descent.</li>
<li><strong>Model Interpretability:</strong> Coefficients represent the change in the log-odds of the outcome for a one-unit change in the predictor.</li>
</ul>

<h2>Step-by-Step Guide to Building Logistic Regression</h2>
<ol>
<li><strong>Data Collection:</strong> Gather the data you will be working with.</li>
<li><strong>Data Preprocessing:</strong>
  <ul>
    <li>Handle missing values.</li>
    <li>Encode categorical variables.</li>
    <li>Scale features.</li>
  </ul>
</li>
<li><strong>Model Specification:</strong> Specify the logistic regression model.

$$\hat{p}=\sigma(\beta_{0}+\beta_{1}X_{1}+\beta_{2}X_{2}+...+\beta_{n}X_{n})$$

</li>
<li><strong>Cost Function Definition:</strong> Define the cost function (log loss).

$$J(\beta)=-\frac{1}{m}\sum^{m}_{i=1}[y_{i}log(\hat{p}_{i})+(1-y_{i})log(1-\hat{p}_{i})]$$

</li>
<li><strong>Gradient Descent:</strong> Derive the gradients and update rules.

$$\frac{dJ(\beta)}{d\beta_{j}}=\frac{1}{m}\sum^{m}_{i=1}(\hat{p}_{i}-y_{i})X_{ij}$$
$$\beta_{j} := \beta_{j}-\alpha\frac{dJ(\beta)}{d\beta_{j}}$$
</li>
<li><strong>Model Training:</strong>
  <ul>
    <li>Initialize coefficients.</li>
    <li>Iterate over the training data to update the coefficients using the gradient descent rule.</li>
  </ul>
</li>
<li><strong>Model Prediction:</strong> Use the trained model to predict probabilities for the new data.

$$\hat{p}=\sigma(\beta_{0}+\beta_{1}X_{1}+\beta_{2}X_{2}+...+\beta_{n}X_{n})$$
</li>
<li><strong>Model Evaluation:</strong>
  <ul>
    <li>Calculate metrics like accuracy, precision, recall, F1 score, and ROC-AUC.</li>

<h1>3. Decision Tree</h1>

<h2>Assumptions</h2>
<ul>
    <li>The relationships between the features and the target variable can be represented by a tree structure.</li>
    <li>The dataset is representative and sufficiently diverse to allow the tree to learn meaningful patterns.</li>
</ul>

<h2>Mathematical Concepts</h2>
<h3>Impurity Measures</h3>
<ul>
    <li><strong>Gini Impurity:</strong> Measures the frequency at which any element of the dataset would be misclassified.
    
$$Gini(D)=1-\sum^{n}_{i=1}p_{i}^2$$
where <i>p<sub>i</sub></i> is the probability of an element being classified into class <i>i</i> .
    </li>
    <li><strong>Entropy:</strong> Measures the amount of randomness or disorder in the dataset.

$$Entropy(D)=-\sum^{n}_{i=1}p_{i}log_{2}(p_{i})$$
where <i>p<sub>i</sub></i> is the probability of an element being classified into class <i>i</i> .
    </li>
</ul>

<h3>Information Gain</h3>
<p>Information gain measures the reduction in impurity or entropy after a dataset is split on an attribute.</p>

$$Information Gain=Entropy(parent)-\sum^{k}_{i=1}\frac{|child_{i}|}{|parent|}Entropy(child_{i})$$
where | <i>child<sub>i</sub></i> | is the number of instances in the <i>i </i> -th subset and | <i>parent</i> | is the total number of instances in the parent set.

<h2>Pros and Cons</h2>
<h3>Pros</h3>
<ul>
    <li>Simple to understand and interpret.</li>
    <li>Requires little data preprocessing.</li>
    <li>Handles both numerical and categorical data.</li>
    <li>Captures non-linear relationships.</li>
</ul>

<h3>Cons</h3>
<ul>
    <li>Prone to overfitting, especially with deep trees.</li>
    <li>Sensitive to variations in data; small changes can result in different trees.</li>
    <li>Biased towards dominant classes.</li>
</ul>

<h2>Techniques to Mitigate Disadvantages</h2>
<ul>
    <li>Pruning: Remove parts of the tree that do not provide significant power to classify instances, to reduce overfitting</li>
    <li>Ensemble Methods: Combine multiple decision trees to improve performance (e.g., Random Forests, Gradient Boosting Trees)</li>
</ul>

<h2>Popular Decision Tree Algorithms</h2>
<ul>
    <li>ID3 (Iterative Dichotomiser 3): Uses entropy and information gain for splitting</li>
    <li>C4.5: Improvement over ID3, handles both continuous and categorical data, and introduces pruning</li>
    <li>CART (Classification and Regression Trees): Uses Gini impurity for classification and MSE for regression</li>
    <li>CHAID (Chi-squared Automatic Interaction Detector): Uses chi-squared tests to find splits</li>
</ul>

<h2>Step-by-Step Guide to Creating a Decision Tree Algorithm from Scratch</h2>
<h3>1. Select the Best Feature to Split</h3>
<p>At each node, evaluate all features to determine the best one to split on based on a chosen criterion (e.g., Gini impurity, entropy).</p>

<h3>2. Determine the Split Point</h3>
<ul>
    <li><strong>Continuous Features:</strong> Sort the data and evaluate possible split points between each pair of consecutive values.</li>
    <li><strong>Categorical Features:</strong> Evaluate each category and possible combinations of categories for splitting.</li>
</ul>

<h3>3. Calculate Impurity/Information Gain for Each Split</h3>
<p>For each possible split, calculate the impurity (e.g., Gini impurity, entropy) for the resulting subsets and choose the split that minimizes impurity or maximizes information gain.</p>

<h3>4. Split the Data</h3>
<p>Divide the dataset into subsets based on the best split point, creating branches for the node.</p>

<h3>5. Repeat Recursively</h3>
<p>For each subset, repeat the process of selecting the best feature and split point, creating new nodes and further dividing the subsets until a stopping condition is met.</p>

<h3>6. Stopping Conditions</h3>
<ul>
    <li>Maximum depth: The maximum depth of the tree is reached.</li>
    <li>Minimum samples per node: A node contains fewer than a specified number of samples.</li>
    <li>Minimum information gain: The information gain from further splits falls below a threshold.</li>
</ul>

<h3>7. Assign Class Labels</h3>
<p>Once the stopping condition is met, assign a class label to the leaf node based on the majority class of the samples in that subset.</p>

<h2>Applying the Decision Tree on Test Data</h2>
<ol>
    <li>Use the fixed tree structure built during training.</li>
    <li>For each test data point, start at the root node and follow the decision rules based on the feature values of the data point.</li>
    <li>Continue following the appropriate branches until reaching a leaf node.</li>
    <li>Assign the class label or regression value of the leaf node to the test data point.</li>
</ol>

<h3>Handling Unseen Feature Values</h3>
<ul>
    <li><strong>Continuous Features:</strong> Any numerical value will be covered by the intervals created by the splits.</li>
    <li><strong>Categorical Features:</strong> If a category was not seen during training, many implementations handle this by sending the data point to the majority class or a default leaf node that covers unseen categories.</li>
</ul>


```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the model
clf = DecisionTreeClassifier(criterion='gini', max_depth=3)
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

# Visualize the tree using matplotlib
plt.figure(figsize=(12,8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, rounded=True)
plt.show()
```


<h1>4. Random Forest</h1>

<h2>1. Introduction to Random Forest</h2>
<p>
    <b>Random Forest</b> is an ensemble learning method that constructs multiple decision trees during training 
    and outputs the class that is the mode of the classes (classification) or mean prediction (regression) of the 
    individual trees. It is known for its simplicity and flexibility, as well as its robustness to overfitting.
</p>

<h2>2. Assumptions</h2>
<p>
    Random Forests are non-parametric and make no strict assumptions about the distribution of the data or the linearity 
    of relationships. They do rely on the assumption that multiple weak learners (individual trees) can be combined to form 
    a strong learner.
</p>

<h2>3. Pros and Cons</h2>
<h3>Pros:</h3>
<ul>
    <li><b>Robustness to Overfitting:</b> By averaging multiple trees, Random Forests reduce the risk of overfitting.</li>
    <li><b>High Accuracy:</b> They generally provide high accuracy and good performance on a variety of tasks.</li>
    <li><b>Handles Missing Values:</b> Can handle missing data well by using median/mode for imputation.</li>
    <li><b>Handles Large Datasets:</b> Efficient on large datasets.</li>
    <li><b>Feature Importance:</b> Provides insights into feature importance, which helps in feature selection.</li>
</ul>
<h3>Cons:</h3>
<ul>
    <li><b>Complexity:</b> The model can become complex and difficult to interpret compared to a single decision tree.</li>
    <li><b>Computationally Intensive:</b> Training can be time-consuming and requires significant computational resources.</li>
    <li><b>Memory Usage:</b> Can consume a lot of memory due to the storage of multiple trees.</li>
</ul>

<h2>4. Combating Disadvantages</h2>
<ul>
    <li><b>Complexity:</b> Use model interpretation techniques like SHAP values or Partial Dependence Plots to understand the model.</li>
    <li><b>Computational Cost:</b> Use techniques like bootstrapping with a smaller subset of data or distributed computing to speed up training.</li>
    <li><b>Memory Usage:</b> Limit the number of trees or use more efficient data structures to store trees.</li>
</ul>

<h2>5. Time Complexity</h2>
<p>
    The time complexity for training a Random Forest is <code>O(n &middot; m &middot; d &middot; log d)</code>, where <code>n</code> is the number of trees, 
    <code>m</code> is the number of samples, and <code>d</code> is the number of features. The prediction time complexity is <code>O(n &middot; log m)</code>.
</p>

<h2>6. Applications</h2>
<ul>
    <li><b>Classification:</b> Email spam detection, disease diagnosis, image classification.</li>
    <li><b>Regression:</b> Stock price prediction, house price prediction, demand forecasting.</li>
    <li><b>Feature Selection:</b> Identifying important features in a dataset.</li>
    <li><b>Anomaly Detection:</b> Identifying outliers in data.</li>
</ul>

<h2>7. Mathematical Concepts</h2>
<ul>
    <li><b>Decision Trees:</b> Random Forests are built using multiple decision trees.</li>
    <li><b>Bootstrap Aggregating (Bagging):</b> Each tree is trained on a different random subset of the training data with replacement.</li>
    <li><b>Random Feature Selection:</b> At each split in the tree, a random subset of features is chosen, which adds an extra layer of randomness and helps in decorrelating the trees.</li>
    <li><b>Voting/Averaging:</b> For classification tasks, the final prediction is made by majority voting, and for regression tasks, the average of the predictions is taken.</li>
</ul>

<h2>8. Creating the Algorithm from Scratch</h2>
<p>
    Here's a high-level overview of how to build a Random Forest algorithm without showing code:
</p>
<ol>
    <li><b>Bootstrap Sampling:</b>
        <ul>
            <li>Randomly sample <code>m</code> data points from the original dataset <code>D</code> with replacement to create <code>n</code> bootstrap samples.</li>
        </ul>
    </li>
    <li><b>Tree Construction:</b>
        <ul>
            <li>For each bootstrap sample, grow a decision tree.</li>
            <li>At each node in the tree, select a random subset of features.</li>
            <li>Split the node based on the feature that provides the best split according to a chosen metric (e.g., Gini impurity for classification, variance reduction for regression).</li>
        </ul>
    </li>
    <li><b>Tree Growth:</b>
        <ul>
            <li>Continue splitting until a stopping criterion is met (e.g., maximum depth, minimum number of samples per leaf).</li>
        </ul>
    </li>
    <li><b>Aggregation:</b>
        <ul>
            <li>For classification, aggregate the predictions from all the trees and take the majority vote.</li>
            <li>For regression, aggregate the predictions from all the trees and take the mean.</li>
        </ul>
    </li>
</ol>

<h2>9. Detailed Steps in Random Forest Algorithm</h2>
<ol>
    <li><b>Initialize Parameters:</b>
        <ul>
            <li>Number of trees (<code>n</code>)</li>
            <li>Number of features to consider at each split (<code>k</code>)</li>
            <li>Stopping criteria (e.g., maximum depth, minimum samples per leaf)</li>
        </ul>
    </li>
    <li><b>For each tree <code>i</code> (from 1 to <code>n</code>):</b>
        <ul>
            <li><b>Sample:</b> Draw a bootstrap sample from the original data.</li>
            <li><b>Tree Building:</b>
                <ul>
                    <li>At each node:</li>
                    <li>Randomly select <code>k</code> features.</li>
                    <li>Determine the best split based on the chosen feature subset.</li>
                    <li>Split the node into two child nodes.</li>
                    <li>Repeat recursively for each child node until stopping criteria are met.</li>
                </ul>
            </li>
        </ul>
    </li>
    <li><b>Predict:</b>
        <ul>
            <li>For each new observation:</li>
            <li>Pass the observation through each tree.</li>
            <li>For classification, collect the votes from each tree and return the majority class.</li>
            <li>For regression, collect the predictions from each tree and return the mean.</li>
        </ul>
    </li>
</ol>

<h2>10. Additional Considerations</h2>
<ul>
    <li><b>Hyperparameters:</b> Number of trees, maximum depth, minimum samples per split/leaf, number of features to consider at each split.</li>
    <li><b>Model Evaluation:</b> Use techniques like cross-validation to evaluate the model's performance.</li>
    <li><b>Feature Importance:</b> Calculate the importance of each feature based on the average decrease in impurity (Gini impurity or variance) across all trees.</li>
</ul>


<h1>5. Support Vector Machine (SVM)</h1>

<h2>1. Introduction to Support Vector Machine (SVM)</h2>
<p><b>Support Vector Machine (SVM)</b> is a supervised machine learning algorithm used for classification and regression tasks. It finds the hyperplane that best separates different classes in the feature space.</p>

<h2>2. Assumptions</h2>
<ul>
    <li><b>Linearly Separable Data:</b> The basic SVM assumes that the data is linearly separable, though kernel methods extend SVM to non-linearly separable data.</li>
    <li><b>Margin Maximization:</b> SVM assumes that the best hyperplane is the one that maximizes the margin between classes.</li>
</ul>

<h2>3. Pros and Cons</h2>
<h3>Pros:</h3>
<ul>
    <li><b>Effective in High Dimensional Spaces:</b> Works well even with a large number of features.</li>
    <li><b>Robust to Overfitting:</b> Especially with the use of appropriate kernels and regularization parameters.</li>
    <li><b>Memory Efficient:</b> Only a subset of training points (support vectors) are used in the decision function.</li>
    <li><b>Versatile:</b> Different kernel functions (linear, polynomial, RBF) can be specified.</li>
</ul>
<h3>Cons:</h3>
<ul>
    <li><b>Complexity in Non-linear Data:</b> Performance can degrade with non-linearly separable data unless using appropriate kernels.</li>
    <li><b>Training Time:</b> Can be computationally intensive for large datasets.</li>
    <li><b>Parameter Tuning:</b> Requires careful tuning of parameters like C (regularization) and kernel parameters.</li>
</ul>

<h2>4. Combating Disadvantages</h2>
<ul>
    <li><b>Non-linear Data:</b> Use kernel tricks (e.g., RBF, polynomial) to handle non-linear relationships.</li>
    <li><b>Large Datasets:</b> Use approximate methods like SVMlight or reduce the dataset size.</li>
    <li><b>Parameter Tuning:</b> Use cross-validation to find optimal parameters.</li>
</ul>

<h2>5. Time Complexity</h2>
<p><b>Training Complexity:</b> <code>O(n<sup>2</sup> &middot; f)</code> for simple SVM, where <code>n</code> is the number of training samples and <code>f</code> is the number of features.<br>
    <b>Prediction Complexity:</b> <code>O(n<sub>s</sub> &middot; f)</code>, where <code>n<sub>s</sub></code> is the number of support vectors.</p>

<h2>6. Applications</h2>
<ul>
    <li><b>Classification:</b> Text categorization, image classification, bioinformatics.</li>
    <li><b>Regression:</b> Predicting continuous values, like stock prices.</li>
    <li><b>Outlier Detection:</b> Identifying unusual data points.</li>
</ul>

<h2>7. Mathematical Concepts</h2>
<ul>
    <li><b>Hyperplane:</b> A decision boundary that separates different classes.</li>
    <li><b>Support Vectors:</b> Data points closest to the hyperplane, influencing its position.</li>
    <li><b>Margin:</b> The distance between the hyperplane and the nearest data points from either class. SVM aims to maximize this margin.</li>
    <li><b>Kernels:</b> Functions that transform the data into a higher-dimensional space to handle non-linear separation (e.g., linear, polynomial, RBF).</li>
</ul>

<h1>Step-by-Step Guide to Creating an SVM Algorithm</h1>
<h2>Step 1: Understand the Problem and Prepare the Data</h2>
<ol>
    <li><b>Problem Definition:</b>
        <ul>
            <li>Determine whether the task is classification or regression.</li>
            <li>Identify the features (inputs) and the target variable (output).</li>
        </ul>
    </li>
    <li><b>Data Collection:</b>
        <ul>
            <li>Collect the dataset relevant to the problem.</li>
        </ul>
    </li>
    <li><b>Data Preprocessing:</b>
        <ul>
            <li>Normalization/Standardization: Scale the features so they have similar ranges.</li>
            <li>Data Cleaning: Handle missing values, outliers, and irrelevant features.</li>
            <li>Data Splitting: Split the dataset into training and testing sets.</li>
        </ul>
    </li>
</ol>

<h2>Step 2: Choose the SVM Type</h2>
<ol>
    <li><b>Linear SVM:</b> Use if the data is linearly separable.</li>
    <li><b>Non-Linear SVM:</b> Use if the data is not linearly separable. Choose an appropriate kernel (e.g., RBF, polynomial).</li>
</ol>

<h2>Step 3: Define the Objective Function and Constraints</h2>
<ol>
    <li><b>Linear SVM:</b>
        <ul>
            <li>Objective: Maximize the margin between the two classes.</li>
            <li>Formulate the primal optimization problem:</li>
            
$$min_{\mathbf{w}, b} \frac{1}{2} \|\mathbf{w}\|^2$$


<li>subject to:</li> 

$$y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1, \quad \forall i$$

</ul>
</li>
<li><b>Soft Margin SVM:</b>
    <ul>
        <li>Allow some misclassification using slack variables <b>x<sub>i</sub></b>:</li>
        
$$min_{\mathbf{w}, b, \xi} \frac{1}{2} \|\mathbf{w}\|^2 + C \sum_{i=1}^n \xi_i$$

<li>subject to:</li>

$$y_i (\mathbf{w} \cdot \mathbf{x}_i + b) \geq 1 - \xi_i, \quad \forall i$$

</ul>
</li>
<li><b>Kernel SVM:</b>
    <ul>
        <li>Use kernel functions to map data to higher dimensions.</li>
        <li>Modify the objective function to include the kernel:</li>

$$K(\mathbf{x}_i, \mathbf{x}_j) = \phi(\mathbf{x}_i) \cdot \phi(\mathbf{x}_j)$$
    
</ul>
</li>
</ol>

<h2>Step 4: Solve the Dual Problem Using Lagrange Multipliers</h2>
<ol>
    <li><b>Formulate the Lagrangian:</b>
        <ul>
            <li>Introduce Lagrange multipliers <b>&alpha;<sub>i</sub></b> for constraints.</li>
            <li>Lagrangian function:</li>
    
$$L(\mathbf{w}, b, \alpha) = \frac{1}{2} \|\mathbf{w}\|^2 - \sum_{i=1}^n \alpha_i \left[ y_i (\mathbf{w} \cdot \mathbf{x}_i + b) - 1 \right]$$

</ul>
    </li>
    <li><b>Convert to Dual Problem:</b>
        <ul>
            <li>Solve for <b>w</b> and <b>b</b> in terms of <b>&alpha;<sub>i</sub></b>:</li>
    
$$\mathbf{w} = \sum_{i=1}^n \alpha_i y_i \mathbf{x}_i$$

<li>Dual problem:</li>

$$max_{\alpha} \sum_{i=1}^n \alpha_i - \frac{1}{2} \sum_{i=1}^n \sum_{j=1}^n \alpha_i \alpha_j y_i y_j K(\mathbf{x}_i, \mathbf{x}_j)$$

<li>subject to:</li>
            
$$\sum_{i=1}^n \alpha_i y_i = 0 \quad \text{and} \quad \alpha_i \geq 0, \quad \forall i$$

</ul>
    </li>
</ol>

<h2>Step 5: Use Quadratic Programming to Solve the Optimization Problem</h2>
<ol>
    <li><b>Quadratic Programming Solver:</b>
        <ul>
            <li>Utilize a quadratic programming solver to find the optimal <b>&alpha;<sub>i</sub></b>.</li>
        </ul>
    </li>
    <li><b>Support Vectors:</b>
        <ul>
            <li>Identify the support vectors as those data points for which <b>&alpha;<sub>i</sub></b> > 0.</li>
        </ul>
    </li>
</ol>

<h2>Step 6: Construct the Decision Boundary</h2>
<ol>
    <li><b>Weight Vector:</b>
        <ul>
            <li>Compute the weight vector <b>w</b>:</li>

$$\mathbf{w} = \sum_{i=1}^n \alpha_i y_i \mathbf{x}_i$$

</ul>
    </li>
    <li><b>Bias Term:</b>
        <ul>
            <li>Compute the bias term <b>b</b> using any support vector <b>x<sub>i</sub></b>:</li>

$$b = y_i - \sum_{j=1}^n \alpha_j y_j K(\mathbf{x}_j, \mathbf{x}_i)$$

</ul>
    </li>
</ol>

<h2>Step 7: Make Predictions</h2>
<ol>
    <li><b>Decision Function:</b>
        <ul>
            <li>For a new data point <b>x</b>, calculate:</li>

$$f(\mathbf{x}) = \text{sign}\left( \sum_{i=1}^n \alpha_i y_i K(\mathbf{x}_i, \mathbf{x}) + b \right)$$

</ul>
    </li>
    <li><b>Class Label:</b>
        <ul>
            <li>Assign the class label based on the sign of the decision function.</li>
        </ul>
    </li>
</ol>

<h2>Step 8: Model Evaluation</h2>
<ol>
    <li><b>Performance Metrics:</b>
        <ul>
            <li>Evaluate the model using metrics like accuracy, precision, recall, F1 score, etc., on the test data.</li>
        </ul>
    </li>
    <li><b>Cross-Validation:</b>
        <ul>
            <li>Perform cross-validation to ensure the model's robustness and generalization.</li>
        </ul>
    </li>
</ol>

<h2>Step 9: Parameter Tuning</h2>
<ol>
    <li><b>Grid Search/Random Search:</b>
        <ul>
            <li>Use grid search or random search to find the optimal values for hyperparameters like <b>C</b>, <b>&gamma;</b>, and kernel parameters.</li>
        </ul>
    </li>
    <li><b>Cross-Validation:</b>
        <ul>
            <li>Use cross-validation to validate the chosen hyperparameters.</li>
        </ul>
    </li>
</ol>

<h2>Summary</h2>
<ol>
    <li><b>Prepare Data:</b> Collect, clean, and split data.</li>
    <li><b>Choose SVM Type:</b> Linear or non-linear.</li>
    <li><b>Define Objective and Constraints:</b> Formulate the optimization problem.</li>
    <li><b>Solve Dual Problem:</b> Use Lagrange multipliers.</li>
    <li><b>Quadratic Programming:</b> Solve for optimal <b>&alpha;<sub>i</sub></b>.</li>
    <li><b>Construct Decision Boundary:</b> Compute <b>w</b> and <b>b</b>.</li>
    <li><b>Make Predictions:</b> Use the decision function.</li>
    <li><b>Evaluate Model:</b> Use performance metrics and cross-validation.</li>
    <li><b>Parameter Tuning:</b> Optimize hyperparameters.</li>
</ol>



<h1>6. K-Nearest Neighbors (KNN)</h1>

<h3>General Description and Application</h3>
<p>K-Nearest Neighbors (KNN) is a simple, instance-based learning algorithm used for both classification and regression tasks. It classifies a data point based on how its neighbors are classified. In regression, it predicts the value of a data point based on the values of its nearest neighbors. KNN is non-parametric, meaning it does not make any assumptions about the underlying data distribution.</p>

<h4>Applications:</h4>
<ul>
    <li>Pattern recognition</li>
    <li>Image classification</li>
    <li>Handwriting detection</li>
    <li>Recommendation systems</li>
    <li>Anomaly detection</li>
</ul>

<h3>Assumptions</h3>
<ul>
    <li>Similar data points are near each other in the feature space.</li>
    <li>The dataset is labeled, meaning each instance has a known output.</li>
</ul>

<h3>Mathematical Foundation</h3>

<h4>1. Distance Metrics:</h4>
<p>KNN relies on a distance metric to find the nearest neighbors. The most common distance metric used is the Euclidean distance.</p>
<div class="equation">
    <p><strong>Euclidean Distance:</strong></p>
    
$$d(p,q)=\sqrt{\sum^{n}_{i=1}(p_{i}-q_{i})^{2}}$$
<p>where <code>p</code> and <code>q</code> are two points in <code>n</code>-dimensional space.</p>
</div>

<h4>2. Finding Neighbors:</h4>
<p>For a given data point, compute the distance to all other points in the dataset and select the <code>k</code> points that have the smallest distances.</p>

<h4>3. Classification:</h4>
<p>The class of the new data point is determined by the majority vote of its <code>k</code> nearest neighbors.</p>
<div class="equation">

$$\hat{y}=mode(y_{1},y_{2},...,y_{n})$$
<p>where <code>y<sub>i</sub></code> are the labels of the <code>k</code> nearest neighbors.</p>
</div>

<h4>4. Regression:</h4>
<p>The value of the new data point is the average of the values of its <code>k</code> nearest neighbors.</p>
<div class="equation">

$$\hat{y}=\frac{1}{k}\sum^{k}_{i=1}y_{i}$$
</div>

<h3>Pros and Cons</h3>

<h4>Pros:</h4>
<ul>
    <li>Simple and easy to understand.</li>
    <li>No training phase (lazy learning).</li>
    <li>Effective with a small number of features.</li>
</ul>

<h4>Cons:</h4>
<ul>
    <li>Computationally expensive during prediction.</li>
    <li>Requires a large amount of memory.</li>
    <li>Sensitive to the choice of <code>k</code>.</li>
    <li>Sensitive to irrelevant or redundant features.</li>
</ul>

<h3>How to Combat Cons</h3>
<ul>
    <li><strong>Dimensionality Reduction:</strong> Use techniques like PCA (Principal Component Analysis) to reduce the number of features.</li>
    <li><strong>Feature Scaling:</strong> Normalize or standardize features to ensure that distances are computed on the same scale.</li>
    <li><strong>Optimal <code>k</code> Selection:</strong> Use cross-validation to find the optimal number of neighbors.</li>
    <li><strong>Efficient Algorithms:</strong> Use KD-Trees or Ball Trees for faster nearest neighbor searches.</li>
</ul>

<h3>Time Complexity</h3>
<ul>
    <li><strong>Training:</strong> <code>O(1)</code> - No training phase.</li>
    <li><strong>Prediction:</strong> <code>O(n &middot; d)</code> - For each query point, compute distance to all <code>n</code> points in <code>d</code>-dimensional space.</li>
</ul>

<h3>Step-by-Step Guide to Create the Algorithm from Scratch</h3>
<ol>
    <li><strong>Collect Data:</strong> Gather a labeled dataset.</li>
    <li><strong>Preprocess Data:</strong>
        <ul>
            <li>Handle missing values.</li>
            <li>Normalize/standardize features.</li>
        </ul>
    </li>
    <li><strong>Choose the Distance Metric:</strong>
        <ul>
            <li>Commonly used is Euclidean distance.</li>
            <li>Other options include Manhattan distance, Minkowski distance, etc.</li>
        </ul>
    </li>
    <li><strong>Choose the Number of Neighbors <code>k</code>:</strong>
        <ul>
            <li>Use cross-validation to select the optimal <code>k</code>.</li>
        </ul>
    </li>
    <li><strong>Implement the Distance Calculation:</strong> Compute the distance between the new data point and all points in the dataset.</li>
    <li><strong>Find Nearest Neighbors:</strong> Sort the distances and select the <code>k</code> smallest distances.</li>
    <li><strong>Classification/Regression:</strong>
        <ul>
            <li>For classification, perform a majority vote on the labels of the <code>k</code> nearest neighbors.</li>
            <li>For regression, calculate the mean of the values of the <code>k</code> nearest neighbors.</li>
        </ul>
    </li>
    <li><strong>Make Predictions:</strong> Use the majority vote or mean value to assign a label or value to the new data point.</li>
</ol>


<h1>7. Naive Bayes</h1>

<h3>General Description and Application</h3>
<p>Naive Bayes is a probabilistic machine learning algorithm used for classification tasks. It is based on Bayes' Theorem with the "naive" assumption of conditional independence between every pair of features given the class label. Despite its simplicity, it performs surprisingly well for a variety of applications.</p>

<h4>Applications:</h4>
<ul>
    <li>Spam detection</li>
    <li>Sentiment analysis</li>
    <li>Document classification</li>
    <li>Medical diagnosis</li>
    <li>Recommendation systems</li>
</ul>

<h3>Assumptions</h3>
<ul>
    <li>Features are conditionally independent given the class label.</li>
    <li>All features contribute equally and independently to the probability.</li>
</ul>

<h3>Mathematical Foundation</h3>

<h4>1. Bayes' Theorem:</h4>
<p>Bayes' Theorem calculates the posterior probability <code>P(C|X)</code> of class <code>C</code> given the feature vector <code>X</code>.</p>

$$P(C|X) = \frac{P(X|C) \cdot P(C)}{P(X)}$$


<h4>2. Naive Assumption:</h4>
<p>Assuming the features are conditionally independent given the class label:</p>

$$P(X|C) = \prod_{i=1}^{n} P(x_i|C)$$

<h4>3. Classification Rule:</h4>
<p>The predicted class <code>Ĉ</code> is the one that maximizes the posterior probability.</p>

$$\hat{C} = \arg\max_{C} P(C|X)$$

<p>Simplifying using Bayes' Theorem and dropping the denominator (as it is the same for all classes):</p>

$$\hat{C} = \arg\max_{C} P(C) \prod_{i=1}^{n} P(x_i|C)$$


<h4>4. Probability Estimation:</h4>
<ul>
    <li><code>P(C)</code> is estimated as the fraction of instances of class <code>C</code> in the training set.</li>
    <li><code>P(x<sub>i</sub>|C)</code> is estimated using different techniques based on the feature type (e.g., Gaussian for continuous features, multinomial for categorical features).</li>
</ul>

<h3>Pros and Cons</h3>

<h4>Pros:</h4>
<ul>
    <li>Simple and easy to implement.</li>
    <li>Efficient with respect to both space and time.</li>
    <li>Works well with high-dimensional data.</li>
    <li>Requires a small amount of training data.</li>
</ul>

<h4>Cons:</h4>
<ul>
    <li>Strong assumption of feature independence, which is rarely true in real-world data.</li>
    <li>Poor performance with correlated features.</li>
    <li>Zero probability problem for unseen feature-class combinations.</li>
</ul>

<h3>How to Combat Cons</h3>
<ul>
    <li><strong>Feature Engineering:</strong> Reduce feature correlation through techniques like PCA.</li>
    <li><strong>Smoothing Techniques:</strong> Use Laplace smoothing to handle zero probabilities:</li>
    
$$P(x_i|C) = \frac{\text{count}(x_i, C) + 1}{\text{count}(C) + k}$$
where <b>k</b> is the number of possible values of <b>x<sub>i</sub></b>
    <li><strong>Ensemble Methods:</strong> Combine Naive Bayes with other algorithms in an ensemble to mitigate its weaknesses.</li>
</ul>

<h3>Time Complexity</h3>
<ul>
    <li><strong>Training:</strong> <code>O(n &middot; d)</code> - Linear with respect to the number of instances <code>n</code> and features <code>d</code>.</li>
    <li><strong>Prediction:</strong> <code>O(d)</code> - Linear with respect to the number of features.</li>
</ul>

<h3>Step-by-Step Guide to Create the Algorithm from Scratch</h3>
<ol>
    <li><strong>Collect Data:</strong> Gather a labeled dataset.</li>
    <li><strong>Preprocess Data:</strong>
        <ul>
            <li>Handle missing values.</li>
            <li>Convert categorical features to numerical values if necessary.</li>
        </ul>
    </li>
    <li><strong>Calculate Prior Probabilities:</strong>
        <ul>
            <li>Estimate the probability of each class <code>P(C)</code> as the fraction of instances of class <code>C</code> in the training set.</li>
        </ul>
    </li>
    <li><strong>Calculate Likelihood Probabilities:</strong>
        <ul>
            <li>Estimate the probability of each feature given the class <code>P(x<sub>i</sub>|C)</code>. For continuous features, assume a Gaussian distribution and estimate the mean and variance for each feature-class combination.</li>
        </ul>
    </li>
    <li><strong>Apply Smoothing:</strong>
        <ul>
            <li>Apply Laplace smoothing to avoid zero probabilities for unseen feature-class combinations.</li>
        </ul>
    </li>
    <li><strong>Calculate Posterior Probabilities:</strong>
        <ul>
            <li>For a given feature vector <code>X</code>, compute the posterior probability for each class using Bayes' Theorem and the naive assumption.</li>
        </ul>
    </li>
    <li><strong>Make Predictions:</strong>
        <ul>
            <li>Choose the class with the highest posterior probability as the predicted class for the given feature vector.</li>
        </ul>
    </li>
    <li><strong>Evaluation:</strong>
        <ul>
            <li>Assess the performance of the model using metrics like accuracy, precision, recall, and F1-score.</li>
        </ul>
    </li>
</ol>



<h1>8. Gradient Boosting Machine (GBM)</h1>
    
<h3>General Description and Application</h3>
<p>Gradient Boosting Machine (GBM) is an ensemble learning technique used for both classification and regression tasks. It builds the model in a stage-wise fashion by training each new model to correct the errors made by the previous models. GBM combines the strengths of multiple weak learners, usually decision trees, to form a strong learner.</p>

<h4>Applications:</h4>
<ul>
    <li>Classification tasks (e.g., spam detection, image recognition)</li>
    <li>Regression tasks (e.g., predicting housing prices, stock market forecasting)</li>
    <li>Ranking tasks (e.g., search engine results ranking)</li>
    <li>Anomaly detection</li>
</ul>

<h3>Assumptions</h3>
<ul>
    <li>The base learners (weak learners) can be combined to improve overall model performance.</li>
    <li>The errors of the combined model can be reduced by focusing on the mistakes of the previous models.</li>
</ul>

<h3>Mathematical Foundation</h3>

<h4>1. Loss Function:</h4>
<p>GBM aims to minimize a loss function <code>L(y, F(x))</code>, where <code>y</code> is the true value, and <code>F(x)</code> is the predicted value.</p>

<h4>2. Additive Model:</h4>
<p>GBM builds the model in an additive manner:</p>

$$F_m(x) = F_{m-1}(x) + \nu \cdot h_m(x)$$

<p>where <code>F<sub>m</sub>(x)</code> is the ensemble model at iteration <code>m</code>, <code>h<sub>m</sub>(x)</code> is the new weak learner, and <code>&nu;</code> is the learning rate.</p>

<h4>3. Gradient Descent:</h4>
<p>Each new weak learner <code>h<sub>m</sub>(x)</code> is trained to minimize the loss function by fitting to the negative gradient of the loss function at the previous iteration:</p>

$$h_m(x) = -\nabla_{F_{m-1}} L(y, F_{m-1}(x))$$


<h4>4. Update Rule:</h4>
<p>The model is updated by adding the new weak learner weighted by the learning rate:</p>

$$F_m(x) = F_{m-1}(x) + \nu \cdot h_m(x)$$


<h3>Pros and Cons</h3>

<h4>Pros:</h4>
<ul>
    <li>High prediction accuracy.</li>
    <li>Can handle various types of data (numerical, categorical, etc.).</li>
    <li>Robust to overfitting if parameters are properly tuned.</li>
</ul>

<h4>Cons:</h4>
<ul>
    <li>Computationally expensive and time-consuming to train.</li>
    <li>Sensitive to hyperparameters (e.g., learning rate, number of trees).</li>
    <li>Requires careful tuning to achieve optimal performance.</li>
</ul>

<h3>How to Combat Cons</h3>
<ul>
    <li><strong>Regularization:</strong> Use techniques like shrinkage (learning rate), subsampling, and tree pruning to prevent overfitting.</li>
    <li><strong>Early Stopping:</strong> Monitor performance on a validation set and stop training when the performance stops improving.</li>
    <li><strong>Hyperparameter Tuning:</strong> Use cross-validation to find the best set of hyperparameters.</li>
</ul>

<h3>Time Complexity</h3>
<ul>
    <li><strong>Training:</strong> <code>O(M &middot; n &middot; d)</code> - Linear with respect to the number of boosting rounds <code>M</code>, the number of instances <code>n</code>, and the number of features <code>d</code>.</li>
    <li><strong>Prediction:</strong> <code>O(M &middot; T)</code> - Linear with respect to the number of boosting rounds <code>M</code> and the number of trees <code>T</code>.</li>
</ul>

    <li><strong>Preprocess Data:</strong>

        </ul>
    </li>
    <li><strong>Initialize Model:</strong> Start with an initial model, typically a constant value (e.g., mean of the target variable).</li>
    <li><strong>Iterative Process:</strong> Repeat for <code>M</code> boosting rounds:
        <ul>
            <li>Compute the negative gradient of the loss function with respect to the current model's predictions.</li>

            <li>Update the model by adding the new weak learner weighted by the learning rate.</li>

<h1>9. XGBoost</h1>

<h3>General Description and Application</h3>
<p>XGBoost (Extreme Gradient Boosting) is an optimized distributed gradient boosting library designed to be highly efficient, flexible, and portable. It implements machine learning algorithms under the Gradient Boosting framework. XGBoost provides a parallel tree boosting that solves many data science problems in a fast and accurate way.</p>

<h4>Applications:</h4>
<ul>
    <li>Classification tasks (e.g., spam detection, image recognition)</li>
    <li>Regression tasks (e.g., predicting housing prices, stock market forecasting)</li>
    <li>Ranking tasks (e.g., search engine results ranking)</li>
    <li>Anomaly detection</li>
</ul>

<h3>Assumptions</h3>
<ul>
    <li>The base learners (weak learners) can be combined to improve overall model performance.</li>
    <li>The errors of the combined model can be reduced by focusing on the mistakes of the previous models.</li>
</ul>

<h3>Mathematical Foundation</h3>

<h4>1. Objective Function:</h4>
<p>XGBoost aims to minimize a regularized objective function that includes a convex loss function and a regularization term.</p>

$$Obj(\theta) = \sum_{i} L(y_i, \hat{y}_i) + \sum_{k} \Omega(f_k)$$

<p>where <code>L</code> is the loss function, <code>&Omega;</code> is the regularization term, <code>y<sub>i</sub></code> is the true value, <code>ŷ<sub>i</sub></code> is the predicted value, and <code>f<sub>k</sub></code> is the <code>k</code>-th tree.</p>

<h4>2. Additive Model:</h4>
<p>XGBoost builds the model in an additive manner:</p>

$$\hat{y}_i^{(t)} = \hat{y}_i^{(t-1)} + f_t(x_i)$$

<p>where <code>ŷ<sub>i</sub><sup>(t)</sup></code> is the prediction of the <code>i</code>-th instance at iteration <code>t</code> and <code>f<sub>t</sub></code> is the <code>t</code>-th tree.</p>

<h4>3. Gradient and Hessian:</h4>
<p>For a given loss function, the gradient <code>g<sub>i</sub></code> and the Hessian <code>h<sub>i</sub></code> (second derivative) are calculated:</p>

$$g_{i} = \frac{\partial L(y_i, \hat{y}_i)}{\partial \hat{y}_i}, \quad h_i = \frac{\partial^2 L(y_i, \hat{y}_i)}{\partial \hat{y}_i^2}$$


<h4>4. Tree Structure:</h4>
<p>XGBoost uses decision trees as base learners. Each tree is built to fit the residual errors of the previous trees.</p>

<h4>5. Regularization:</h4>
<p>Regularization is applied to the trees to prevent overfitting:</p>

$$\Omega(f) = \gamma T + \frac{1}{2} \lambda \sum_{j=1}^{T} w_j^2$$

<p>where <code>&gamma;</code> is the penalty term for the number of leaves <code>T</code> in the tree, and <code>&lambda;</code> is the L2 regularization term on the leaf weights <code>w<sub>j</sub></code>.</p>

<h4>6. Split Finding:</h4>
<p>The best split is found by maximizing the gain from a split:</p>

$$Gain = \frac{1}{2} \left[ \frac{G_L^2}{H_L + \lambda} + \frac{G_R^2}{H_R + \lambda} - \frac{(G_L + G_R)^2}{H_L + H_R + \lambda} \right] - \gamma$$

<p>where <code>G</code> and <code>H</code> are the sums of the gradients and Hessians, respectively, for the left and right splits.</p>

<h3>Pros and Cons</h3>

<h4>Pros:</h4>
<ul>
    <li>High prediction accuracy.</li>
    <li>Can handle missing values internally.</li>
    <li>Supports parallel and distributed computing.</li>
    <li>Robust to overfitting with proper regularization.</li>
    <li>Highly customizable and flexible with various hyperparameters.</li>
</ul>

<h4>Cons:</h4>
<ul>
    <li>Computationally intensive and requires more memory.</li>
    <li>Sensitive to hyperparameters.</li>
    <li>Requires careful tuning to achieve optimal performance.</li>
</ul>

<h3>How to Combat Cons</h3>
<ul>
    <li><strong>Regularization:</strong> Use regularization techniques provided by XGBoost, such as L1 and L2 regularization, to prevent overfitting.</li>
    <li><strong>Early Stopping:</strong> Monitor performance on a validation set and stop training when the performance stops improving.</li>
    <li><strong>Hyperparameter Tuning:</strong> Use cross-validation and grid search to find the best set of hyperparameters.</li>
    <li><strong>Distributed Computing:</strong> Leverage XGBoost's support for distributed computing to handle large datasets efficiently.</li>
</ul>

<h3>Time Complexity</h3>
<ul>
    <li><strong>Training:</strong> <code>O(M &middot; n &middot; log(n))</code> - Linear with respect to the number of boosting rounds <code>M</code>, the number of instances <code>n</code>, and the logarithm of <code>n</code>.</li>
    <li><strong>Prediction:</strong> <code>O(M &middot; T)</code> - Linear with respect to the number of boosting rounds <code>M</code> and the number of trees <code>T</code>.</li>
</ul>

<h3>Step-by-Step Guide to Create the Algorithm from Scratch</h3>
<ol>
    <li><strong>Collect Data:</strong> Gather a labeled dataset.</li>
    <li><strong>Preprocess Data:</strong>
        <ul>
            <li>Handle missing values.</li>
            <li>Encode categorical variables.</li>
            <li>Normalize/standardize features if necessary.</li>
        </ul>
    </li>
    <li><strong>Initialize Model:</strong> Start with an initial model, typically a constant value (e.g., mean of the target variable).</li>
    <li><strong>Iterative Process:</strong> Repeat for <code>M</code> boosting rounds:
        <ul>
            <li>Compute the gradient <code>g<sub>i</sub></code> and Hessian <code>h<sub>i</sub></code> for each instance.</li>
            <li>Fit a weak learner (e.g., decision tree) to the negative gradient.</li>
            <li>Update the model by adding the new weak learner weighted by the learning rate.</li>
        </ul>
    </li>
    <li><strong>Regularization:</strong> Apply regularization terms to the trees to prevent overfitting.</li>
    <li><strong>Split Finding:</strong> Determine the best splits by maximizing the gain.</li>
    <li><strong>Update Model:</strong> Update the ensemble model with the newly trained tree.</li>
    <li><strong>Prediction:</strong> For a new instance, sum the predictions of all weak learners to get the final prediction.</li>
    <li><strong>Evaluation:</strong> Assess the performance of the model using appropriate metrics (e.g., accuracy for classification, RMSE for regression).</li>
</ol>



<h1>10. AdaBoost</h1>

<h3>General Description and Application</h3>
<p>AdaBoost (Adaptive Boosting) is an ensemble learning method primarily used for binary classification. It combines multiple weak learners, typically decision stumps (trees with a single split), to create a strong classifier. AdaBoost adjusts the weights of misclassified instances to focus more on hard-to-classify examples in subsequent iterations.</p>

<h4>Applications:</h4>
<ul>
    <li>Binary classification tasks (e.g., spam detection, fraud detection)</li>
    <li>Multi-class classification (using one-vs-one or one-vs-all approaches)</li>
    <li>Face detection</li>
    <li>Object recognition</li>
</ul>

<h3>Assumptions</h3>
<ul>
    <li>The weak learners perform better than random guessing.</li>
    <li>The final model is a linear combination of the weak learners.</li>
</ul>

<h3>Mathematical Foundation</h3>

<h4>1. Initialization:</h4>
<p>Initialize the weights of the training instances. Each instance is assigned an equal weight initially:</p>

$$w_i^{(1)} = \frac{1}{n}$$

<p>where <code>n</code> is the number of training instances.</p>

<h4>2. Iterative Process:</h4>
<p>For <code>m = 1</code> to <code>M</code> (number of boosting rounds):</p>
<ul>
    <li>Train a weak learner <code>h<sub>m</sub>(x)</code> on the weighted training data.</li>
    <li>Calculate the error of the weak learner:</li>
    
$$\epsilon_m = \frac{\sum_{i=1}^{n} w_i^{(m)} \cdot I(y_i \neq h_m(x_i))}{\sum_{i=1}^{n} w_i^{(m)}}$$
    
<p>where <code>I(&middot;)</code> is the indicator function that equals 1 if the condition is true and 0 otherwise.</p>
    <li>Compute the weight of the weak learner:</li>

$$\alpha = \frac{1}{2} \ln \left( \frac{1 - \epsilon_m}{\epsilon_m} \right)$$
    
<li>Update the weights of the training instances:</li>
    
$$w_i^{(m+1)} = w_i^{(m)} \cdot \exp \left( -\alpha_m \cdot y_i \cdot h_m(x_i) \right)$$

<li>Normalize the weights so that they sum to 1:</li>

$$w_i^{(m+1)} = \frac{w_i^{(m+1)}}{\sum_{j=1}^{n} w_j^{(m+1)}}$$

</ul>

<h4>3. Final Model:</h4>
<p>The final strong classifier is a weighted majority vote of the weak learners:</p>

$$H(x) = \text{sign} \left( \sum_{m=1}^{M} \alpha_m \cdot h_m(x) \right)$$


<h3>Pros and Cons</h3>

<h4>Pros:</h4>
<ul>
    <li>Simple and easy to implement.</li>
    <li>Can convert weak learners into a strong classifier.</li>
    <li>Robust to overfitting if not too many rounds are used.</li>
    <li>No parameter tuning is required for the base learners.</li>
</ul>

<h4>Cons:</h4>
<ul>
    <li>Sensitive to noisy data and outliers.</li>
    <li>Can overfit if the number of boosting rounds <code>M</code> is too large.</li>
    <li>Requires high-quality weak learners.</li>
</ul>

<h3>How to Combat Cons</h3>
<ul>
    <li><strong>Early Stopping:</strong> Stop the boosting process early to prevent overfitting.</li>
    <li><strong>Robust Base Learners:</strong> Use base learners that are less sensitive to noise, such as pruned decision trees.</li>
    <li><strong>Regularization:</strong> Introduce regularization techniques like shrinkage (reducing the weight <code>&alpha;<sub>m</sub></code>).</li>
    <li><strong>Outlier Detection:</strong> Preprocess the data to remove or reduce the impact of outliers.</li>
</ul>

<h3>Time Complexity</h3>
<ul>
    <li><strong>Training:</strong> <code>O(M &middot; n &middot; d)</code> - Linear with respect to the number of boosting rounds <code>M</code>, the number of instances <code>n</code>, and the number of features <code>d</code>.</li>
    <li><strong>Prediction:</strong> <code>O(M &middot; d)</code> - Linear with respect to the number of boosting rounds <code>M</code> and the number of features <code>d</code>.</li>
</ul>

<h3>Step-by-Step Guide to Create the Algorithm from Scratch</h3>
<ol>
    <li><strong>Collect Data:</strong> Gather a labeled dataset.</li>
    <li><strong>Preprocess Data:</strong>
        <ul>
            <li>Handle missing values.</li>
            <li>Encode categorical variables.</li>
            <li>Normalize/standardize features if necessary.</li>
        </ul>
    </li>
    <li><strong>Initialize Weights:</strong> Initialize the weights of the training instances equally:</li>
    
$$w_i^{(1)} = \frac{1}{n}$$
    
<li><strong>Iterative Process:</strong> Repeat for <code>M</code> boosting rounds:
        <ul>
            <li>Train a weak learner <code>h<sub>m</sub>(x)</code> on the weighted training data.</li>
            <li>Calculate the error <code>&epsilon;<sub>m</sub></code> of the weak learner.</li>
            <li>Compute the weight <code>&alpha;<sub>m</sub></code> of the weak learner.</li>
            <li>Update the weights <code>w<sub>i<sub><sup>(m+1)</sup></code> of the training instances.</li>
            <li>Normalize the weights.</li>
        </ul>
    </li>
    <li><strong>Final Model:</strong> Combine the weak learners into a final strong classifier:</li>

$$H(x) = \text{sign} \left( \sum_{m=1}^{M} \alpha_m \cdot h_m(x) \right)$$
    
<li><strong>Prediction:</strong> For a new instance, compute the weighted majority vote of the weak learners to get the final prediction.</li>
    <li><strong>Evaluation:</strong> Assess the performance of the model using appropriate metrics (e.g., accuracy for classification).</li>
</ol>
